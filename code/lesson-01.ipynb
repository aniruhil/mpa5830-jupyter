# Welcome to the tidyverse

We are going to start by loading the tidyverse, an idiosyncratic collection of packages that play well together, by design and not by accident! We will then load a particular dataset -- the `diamonds` dataset, and then build a simple `scatterplot` of `price` versus `carat` (the weight of the diamond). 

```r
library(tidyverse)
```

```r
glimpse(diamonds)
```

```r
diamonds %>%
    ggplot() +
    geom_point(aes(x = carat, y = price, color = cut))
```
